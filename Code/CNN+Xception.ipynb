```python
import tensorflow as tf
from tensorflow.keras import layers, models
from tensorflow.keras.applications import Xception
from tensorflow.keras.callbacks import EarlyStopping, ModelCheckpoint
import matplotlib.pyplot as plt

# === CONFIGURATION ===
BATCH_SIZE = 32
IMAGE_DIM = (224, 224)
DATA_PATH = "Dataset"  # Should include subfolders 'real/' and 'fake/'
EPOCHS = 15

# === 1. Dataset Loading ===
train_data = tf.keras.utils.image_dataset_from_directory(
    DATA_PATH,
    validation_split=0.2,
    subset="training",
    seed=42,
    image_size=IMAGE_DIM,
    batch_size=BATCH_SIZE,
    label_mode='binary'  # For binary classification
)

val_data = tf.keras.utils.image_dataset_from_directory(
    DATA_PATH,
    validation_split=0.2,
    subset="validation",
    seed=42,
    image_size=IMAGE_DIM,
    batch_size=BATCH_SIZE,
    label_mode='binary'
)

print("Detected classes:", train_data.class_names)

# === 2. Data Preparation ===
normalizer = layers.Rescaling(1.0 / 255)
AUTOTUNE = tf.data.AUTOTUNE

train_data = (
    train_data
    .map(lambda x, y: (normalizer(x), y))
    .cache()
    .shuffle(1000)
    .prefetch(buffer_size=AUTOTUNE)
)

val_data = (
    val_data
    .map(lambda x, y: (normalizer(x), y))
    .cache()
    .prefetch(buffer_size=AUTOTUNE)
)

# === 3. Load Pretrained Xception Model ===
feature_extractor = Xception(
    input_shape=IMAGE_DIM + (3,),
    include_top=False,
    weights='imagenet'
)
feature_extractor.trainable = False  # Freeze the backbone for now

# === 4. Custom Model Architecture ===
input_tensor = tf.keras.Input(shape=IMAGE_DIM + (3,))
x = feature_extractor(input_tensor, training=False)
x = layers.GlobalAveragePooling2D()(x)
x = layers.Dropout(0.3)(x)
x = layers.Dense(256, activation='relu')(x)
output_tensor = layers.Dense(1, activation='sigmoid')(x)

model = tf.keras.Model(input_tensor, output_tensor)

# === 5. Compilation ===
model.compile(
    optimizer='adam',
    loss='binary_crossentropy',
    metrics=['accuracy']
)

model.summary()

# === 6. Define Callbacks ===
callback_list = [
    EarlyStopping(patience=3, monitor='val_loss', restore_best_weights=True),
    ModelCheckpoint("best_model_xception.keras", save_best_only=True)
]

# === 7. Training the Model ===
history = model.fit(
    train_data,
    validation_data=val_data,
    epochs=EPOCHS,
    callbacks=callback_list
)

# === 8. Evaluation ===
val_loss, val_accuracy = model.evaluate(val_data)
print(f"Validation Accuracy: {val_accuracy:.2f}")

# === 9. Visualization ===
train_acc = history.history['accuracy']
val_acc = history.history['val_accuracy']
train_loss = history.history['loss']
val_loss_curve = history.history['val_loss']

plt.figure(figsize=(12, 5))

plt.subplot(1, 2, 1)
plt.plot(train_acc, label='Training Accuracy')
plt.plot(val_acc, label='Validation Accuracy')
plt.title('Accuracy Over Epochs')
plt.legend()

plt.subplot(1, 2, 2)
plt.plot(train_loss, label='Training Loss')
plt.plot(val_loss_curve, label='Validation Loss')
plt.title('Loss Over Epochs')
plt.legend()

plt.tight_layout()
plt.show()

# Save the model
model.save("xception_fakeface_model.keras")

# === 10. Fine-Tuning the Model ===
# Unfreeze selected layers for fine-tuning
feature_extractor.trainable = True
for layer in feature_extractor.layers[:-10]:  # Keep earlier layers frozen
    layer.trainable = False

# Recompile with a lower learning rate
model.compile(
    optimizer=tf.keras.optimizers.Adam(learning_rate=1e-5),
    loss='binary_crossentropy',
    metrics=['accuracy']
)

# Fine-tune the model
fine_tune_history = model.fit(
    train_data,
    validation_data=val_data,
    epochs=5
)
```

```python
import tensorflow as tf
import numpy as np
import cv2
from google.colab.patches import cv2_imshow
from google.colab import files
from PIL import Image

# Load the pre-trained Xception model
model = tf.keras.models.load_model("xception_fakeface_model.keras")

# Labels for prediction output
labels = ['real', 'fake']

def upload_and_predict():
    # Let user upload an image
    uploaded_file = files.upload()

    # Pick the uploaded file
    for file_name in uploaded_file:
        print(f"Uploaded: {file_name}")
        img = Image.open(file_name)

        # Resize to fit model's input requirement
        img_resized = img.resize((224, 224))
        img_array = np.array(img_resized) / 255.0  # Normalize the image
        img_input = np.expand_dims(img_array, axis=0)

        # Run prediction
        pred = model.predict(img_input)
        pred_class = np.argmax(pred)
        confidence = pred[0][pred_class]
        result = labels[pred_class]

        # Display the result
        if confidence < 0.6:
            print(f"Prediction: Uncertain (Confidence: {confidence*100:.2f}%)")
        else:
            print(f"Prediction: {result.capitalize()} (Confidence: {confidence*100:.2f}%)")

        # Show the image
        cv2_imshow(np.array(img_resized))

# Run the function
upload_and_predict()
```
